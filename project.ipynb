#### Now let's work with the "south belt" of madrid. Make a subsetting that contains the following populations (level5): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆) 

```python
import pandas as pd
import numpy as np
data
data_south_belt1 = data[data['level5']== "Fuenlabrada"]
data_south_belt2 = data[data['level5']== "Leganés"]
data_south_belt3 = data[data['level5']== "Getafe"]
data_south_belt4 = data[data['level5']== "Alcorcón"]
data_south_Total = pd.concat([data_south_belt1,data_south_belt2,data_south_belt3,data_south_belt4], axis=0)
data_south_Total
```

#### Make a bar plot of the median of prices (★★★) 

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

data_south_Total.groupby('level5')['price'].mean().plot(kind = 'bar', title = "precio promedio")
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
import pandas as pd
import numpy as np
%matplotlib inline

data
precio = data.filter(items=['level5','price'])
precio
precio1 = precio.filter(items=['price'])
media_precio = precio1.mean()
desvia = precio1.std()

normalizacion = precio1 - media_precio/desvia
normalizacion

normalizacion.plot(kind = 'hist')
plt.show



```

```python
import pandas as pd
import numpy as np
%matplotlib inline

normalizacion

plt.hist(normalizacion)
plt.title('Histograma del precio normalizado')
plt.xlabel('cantidad')
plt.ylabel('precio')
plt.show()
```
